# Problema 1

Sean las funciones:

\begin{split}
r(x) & = -1,5x^{2} + 11,5x - 15\\
g(x) & = 1,58x^{4} - 19,17x^{3} + 80,92x^{2} - 139,33x + 85
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $0 \leq x \leq 6$, $\hspace{0.3cm}$  $-10 \leq y \leq 10$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.
2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

# Solución

**1)**

```python
def r(x):
    return -1.5*x**2 + 11.5*x -15
def g(x):
    return 1.58*x**4 - 19.17*x**3 + 80.92*x**2 - 139.33*x +85

x = np.arange(-10,10,0.01)

plt.plot(x,r(x), color='blue', label="r(x)")
plt.plot(x,g(x), color='red', label="g(x)")

plt.xlim(0,6)  
plt.ylim(-10,10)
plt.xlabel("valores de x")
plt.ylabel("valores de r(x) y de g(x)")
plt.title("r(x) y g(x) en función d e x")
plt.grid(True)
plt.legend()

plt.show()
```

**2)**

```python
def intersect(x):
   return r(x) - g(x)

x = np.linspace(1,6, 5)

sol = np.around(fsolve(intersect, x), decimals = 1)
var = np.unique(sol)

print(f"las curvas se intersectan en x = {"".join(str(var))}")
```

# Problema 2

Sean las funciones:

\begin{split}
h(x) & = 2(x - 2)^{2}\\
i(x) & = x + 1
\end{split}

<br>

1. Grafique ambas funciones en un mismo gráfico utilizando *Matplotlib*.
2. ¿En qué punto(s) interseca la función $h(x)$ al eje $x$? ¿y la función $i(x)$?
3. ¿Las gráficas de las funciones se intersecan en algún punto?
4. ¿Para qué valores de $x$ ocurre que $h(x) \leq i(x)$?
5. Determine el punto en que $h(x) = 50$.


# Solución

**1)**

```python
def h(x): 
   return 2 * (x - 2) ** 2

def i(x):
   return x + 1

x = np.arange(-5,5,0.01)

plt.plot(x, h(x), label="h(x)")
plt.plot(x, i(x), label="i(x)")
plt.xlim(-2,5)
plt.ylim(-1,10)

plt.xlabel("valores de x")
plt.ylabel("valores de h(x) y i(x)")
plt.legend()

plt.grid(True)
plt.show()
```

**2)**

```python
x = np.linspace(-2.5,10,5)

roots_h = np.around(fsolve(h, x), decimals = 1)
roots_i = np.around(fsolve(i, x), decimals = 1)

sol_h = np.unique(roots_h)
sol_i = np.unique(roots_i)

print(f"La curva h(x) intersecta al eje x en el punto ({sol_h[0]}, {h(sol_h[0])})")
print(f"La curva i(x) intersecta al eje x en el punto ({sol_i[0]}, {i(sol_i[0])})")
```

**3)**

```python
def intersect(x):
   return h(x) - i(x)

x = np.linspace(0,4,2)

roots = np.around(fsolve(intersect, x), decimals = 1)
sol = np.unique(roots)

print(f"Las curvas se intersectan en los puntos ({sol[0]},{h(sol[0])}), ({sol[1]}, {h(sol[1])})")
```

**4)**

$$h(x) \leq i(x) \Leftrightarrow 1 \leq x \leq 3.5

**5)**

```python
def h_50(x):
   return h(x) - 50

x = np.linspace(-5,5,2)

roots = fsolve(h_50, x)

print(f"h(x) = 50 cuando x = {roots[0]:.0f} y x = {roots[1]:.0f}")
```

Sean las funciones:

\begin{split}
w(x) & = 0,6x^{2} + 1,5\\
t(x) & = 0,8x^{3} - 10x + 10
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $-3 \leq x \leq 4$, $\hspace{0.2cm}$ $-4 \leq y \leq 24$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones en un mismo gráfico, indicando unidad de medida. Utilice *Matplotlib*.
2. Determine el intervalo en que $w(x) \geq t(x)$.
3. Encuentre los puntos en que $t(x) = 6$.


# Solución

**1)**

```python
def w(x):
   return 0.6 * x ** 2 + 1.5

def t(x):
   return 0.8 * x ** 3 - 10 * x + 10

x = np.linspace(-5,5)

plt.plot(x, w(x), label = "w(x)")
plt.plot(x, t(x), label = "t(x)")
plt.xlim(-3, 4)
plt.ylim(-4, 24)
plt.xlabel("Valores de x")
plt.ylabel("Valores de w(x) y t(x)")

plt.legend()
plt.show()
```

**2)**

```python
def intersect(x):
   return w(x) - t(x)

x = np.linspace(-4,4,3)

roots = np.around(fsolve(intersect, x), decimals = 2)
sol = np.unique(roots)

print(sol)
```

$$ w(x) \geq t(x) \Leftrightarrow x \in [0.86, 3.47]

**3)**

```python
def t_6(x):
   return t(x) - 6

x = np.linspace(-10,10,3)

roots = np.around(fsolve(t_6, x), decimals = 10)
sol = np.unique(roots)

for i in range(len(sol)):
   print(f"x{i+1} = {sol[i]:.2f}")
```

Donde $x$ corresponde al tiempo transcurrido desde que se inició la campaña (en semanas).

<br>

# Solución

**1)**

**Ingresos**: en la semana 0 el total de ingresos es de 1000 mil pesos, y a medida que avanza el tiempo los ingresos van disminuyendo en 20 mil pesos por semana

$$S_{actual}(x) = 600 \cdot 1,05^{x}$$

<br>

donde $x$ corresponde a la cantidad de casas vendidas durante el mes. Debido al aumento en el costo de la vida, necesita aumentar sus ingresos. $$S_{nuevo}(x) = 500 * 1.07^x$$

# Problema 9

En un laboratorio de biología marina, los científicos están estudiando el crecimiento de dos especies de algas en función de la luz solar recibida. Las funciones $f(x)$ y $g(x)$, que describen la tasa de crecimiento de las algas en milímetros por día (mm/día) son:

<br>

$$f(x) = \sqrt{500 x + 2.000}$$

<br>

$$g(x) = \sqrt{800 x + 1.000}$$

<br>

donde $x$ corresponde a la intensidad de la luz (en lux).

<br>

1. Determina el punto en el que ambas especies de algas tienen la misma tasa de crecimiento. Además, calcula la tasa de crecimiento en ese punto. Visualiza apropiadamente.
2. Considerando lo obtenido en el ítem anterior, analice y compare el crecimiento de las dos especies en función de la intensidad de la luz.

# Solución

**1)**

```python
def f(x):
   return np.sqrt(500 * x + 2000)

def g(x):
   return np.sqrt(800 * x + 1000)

def inter(x):
   return f(x) - g(x)

x = np.arange(0,10)
plt.plot(x, f(x), label="f(x)")
plt.plot(x, g(x), label="g(x)")
plt.plot(3.33, 60.55, '-ro')

plt.title("Relación tasa de crecimiento e intensidad de luz")
plt.xlabel("Intensidad de la luz (lux)")
plt.ylabel("Tasa de crecimiento (mm/dia)")

plt.legend()
plt.grid(True)
plt.show()


x = np.linspace(0,100,1)
roots = fsolve(inter, x)
print(f"punto de intersección: ({roots[0]:.2f}, {f(roots[0]):.2f})")
```

Cuando la intensidad de la luz llega a 3.33 lux aproximadamente, es cuando ambas especies tienen la misma tasa de crecimiento, 60.55 mm/dia aproximadamente

**2)**

Ambas especies aumentan su tasa de crecimiento conforme aumenta la intensidad de la luz. En un principio la primera especie tiene una tasa de crecimiento mayor, pero va aumentando más lentamente que la segunda especie, hasta llegar a un punto de equilibrio cuando hay $3.33$ lux, luego de este punto la segunda especie tiene una tasa de crecimiento mayor.

# Problema 10

En una planta de tratamiento de aguas residuales, los ingenieros están estudiando la eficiencia, medida en porcentaje, de dos diferentes procesos de filtración para eliminar contaminantes. La eficiencia de cada proceso varía en función del tiempo de tratamiento, $t$, medido en horas.

Las funciones que describen la eficiencia de los procesos son las siguientes:

Proceso A:

 $$A(t) = 1,4t^{4} - 2,55t^{3} - 3,37t^{2} + 2,75t + 10$$

Proceso B:

 $$B(t) = t^{4} -2,67t^{3} - t^{2} + 2t + 20$$

 <br>

1. Grafique ambas funciones indicando el nombre de cada eje coordenado, junto con su unidad de medida. Utilice *Matplotlib*.
2. Transcurridas $3$ horas y media, ¿qué porcentaje de eficiencia tiene el proceso A? ¿Y el B?
3. ¿Para cuántas horas transcurridas el proceso A tiene la misma eficiencia que el proceso B? Indique la eficiencia en este punto.
4. Si sabe que los procesos demoran como máximo $5$ horas, analice cuándo es conveniente aplicar el proceso A.

# Solución

**1)**

```python
def a(t):
   return 1.4 * t ** 4 - 2.55 * t ** 3 - 3.37 * t ** 2 + 2.75 * t + 10

def b(t):
   return t ** 4 - 2.67 * t ** 3 - t ** 2 + 2 * t + 20

t = np.arange(0, 10, 0.01)

plt.plot(t, a(t), label="a(t)")
plt.plot(t, b(t), label="b(t)")
plt.ylim(0,100)
plt.xlim(0,5)

plt.title("Relación entre la eficiencia de filtración y tiempo")
plt.xlabel("Tiempo (horas)")
plt.ylabel("Eficiencia (porcentaje)")

plt.legend()
plt.grid(True)
plt.show()
```

**2)**

```python
print(f"porcentaje de eficiencia del proceso A transcurridas 3 horas y media: {a(3.5):.2f}%")
print(f"porcentaje de eficiencia del proceso B transcurridas 3 horas y media: {b(3.5):.2f}%")
```

**3)**

```python
def inter(t):
   return a(t) - b(t)

t = np.linspace(0,24,2)

roots = np.round(fsolve(inter,t),decimals=10)
sol = np.unique(roots)

print(f"Horas transcurridas: {sol[1]:.2f} horas")
print(f"eficiencia: {a(sol[1]):.2f}%")
```

**4)**

El proceso A es más eficiente para tratamientos que duran más de 2.77 horas.

# Problema 11

Sea la función $f$, definida por:

<br>

$$f(x) = \begin{cases}
          & 2 - x  &\text{si } x < 1 \\
          & 3      &\text{si } x\geq 1
\end{cases} $$

<br>

1. Describa el comportamiento del gráfico de la función.
2. ¿Cuál es el valor de $f(x)$ si $x = -5$? ¿Y si $x = 5$?


# Solución

**1)**

```python
def tramo_1(x):
   return 2 - x

def tramo_2(x):
   return 3

tramo_2 = np.vectorize(tramo_2)

x = np.arange(-10, 10, 0.001)
f_x = np.piecewise(x, [x < 1, x >=1], [lambda x: tramo_1(x), lambda x: tramo_2(x)])

plt.plot(x[x < 1], tramo_1(x[x < 1]), color="darkred", label="tramo 1")
plt.plot(x[x >= 1], tramo_2(x[x >= 1]), color="g", label="tramo 2")
plt.show()
```

```python
x_series = np.arange(-10, 10, .01)
rules = [
    (lambda x: 2 - x, x_series < 1),
    (3, x_series >= 1),
]
evaluations, conditions = zip(*rules)

y_series = np.piecewise(x_series, conditions, evaluations)
plt.figure()
marker_style = 'o'
marker_size = 1
plt.grid(True, which='both')
plt.scatter(x_series, y_series, marker=marker_style, s=marker_size)
plt.show()
```

La función es discontinua, para valores de x menores a 1, la función es decreciente, para valores mayores o iguales a 1, la función es constante

**2)**

```python
def f(x):
   return 2 - x if x < 1 else 3

print(f(-5))
print(f(5))
```

$f(x) = 7$ si $x = -5$

$f(x) = 3$ si $x = 5$
